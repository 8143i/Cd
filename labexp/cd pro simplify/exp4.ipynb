```python
e = 'ε'

A, r = input("Enter production: ").split("->")
p = [x.strip() for x in r.split("|")]

# manual grouping
AB = [x for x in p if x.startswith("AB")]
Sg = [x for x in p if x.startswith("S")]
others = [x for x in p if x not in AB+Sg]

print(A.strip(), "->", "AB"+A.strip()+"' | S"+"S' |", " | ".join(others))

# A'
A1 = [x[2:] or e for x in AB]
print(A.strip()+"'", "->", " | ".join(A1))

# S'
S1 = [x[1:] or e for x in Sg]
print("S'", "->", " | ".join(S1))
```
