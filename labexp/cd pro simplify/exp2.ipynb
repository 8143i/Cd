```python
import re

t, eps = 0, 'ε'

def build(exp):
    global t
    if re.fullmatch(r'[a-z]', exp):
        res = [(t, t+1, exp)]; t += 1

    elif re.fullmatch(r'[a-z]\*', exp):
        res = [(t,t+1,eps),(t,t+3,eps),(t+1,t+2,exp[0]),
               (t+2,t+1,eps),(t+2,t+3,eps)]
        t += 3

    elif re.fullmatch(r'[a-z]/[a-z]', exp):
        res = [(t,t+1,eps),(t,t+3,eps),(t+1,t+2,exp[0]),
               (t+3,t+4,exp[2]),(t+2,t+5,eps),(t+4,t+5,eps)]
        t += 5
    else:
        return []
    return res

def show(nfa):
    sym = sorted({x for _,_,x in nfa})
    for s in range(t+1):
        print(s, end="\t")
        for y in sym:
            print([d for a,d,c in nfa if a==s and c==y], end="\t")
        print()
    print("Final:", t)

nfa = []
for x in input("Enter regex: ").split():
    nfa += build(x)

show(nfa)
```
