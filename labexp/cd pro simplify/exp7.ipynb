```python
gram = ["S+S","S*S","S-S","(S)","id"]
inp = "id+id)$"
stack = "$("   # start with '(' already shifted
print(f'{"Stack":<15}|{"Input Buffer":<15}|Parsing Action')
print("-"*50)
while True:
    for g in gram:
        if g in stack:
            stack = stack.replace(g, "S", 1)
            print(f'{stack:<15}|{inp:<15}|Reduce > {g}')
            break
    else:
        stack += inp[0]
        inp = inp[1:]
        print(f'{stack:<15}|{inp:<15}|Shift')
    if stack == "$S" and inp == "$":
        print(f'{stack:<15}|{inp:<15}|Accepted')
        break
```
