```python
ops = set('+-*/()')
pri = {'+':1,'-':1,'*':2,'/':2}

def post(s):
    st, out = [], ''
    for c in s:
        if c not in ops: out += c
        elif c == '(': st.append(c)
        elif c == ')':
            while st[-1] != '(': out += st.pop()
            st.pop()
        else:
            while st and st[-1] != '(' and pri[c] <= pri[st[-1]]:
                out += st.pop()
            st.append(c)
    return out + ''.join(st[::-1])

def tac(p):
    st, t = [], 1
    for c in p:
        if c not in ops: st.append(c)
        else:
            b, a = st.pop(), st.pop()
            print(f"t{t} = {a} {c} {b}")
            st.append(f"t{t}"); t += 1

s = input("Enter expr: ")
p = post(s)
print("Postfix:", p)
tac(p)
```
