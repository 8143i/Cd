```python
p={}
for _ in range(int(input("No of production: "))):
    A,r=input().split("->"); p[A.strip()]=[x.strip() for x in r.split("|")]

F, Fo = {}, {}

def first(A):
    if A not in p: return {A}
    if A in F: return F[A]
    F[A]=set()
    for pr in p[A]:
        for c in pr:
            if not c.isupper(): F[A].add(c); break
            F[A]|=first(c)-{'e'}
            if 'e' not in first(c): break
        else: F[A].add('e')
    return F[A]

def follow(A):
    if A in Fo: return Fo[A]
    Fo[A]={'$'} if A==list(p)[0] else set()
    for X in p:
        for pr in p[X]:
            for i,c in enumerate(pr):
                if c==A:
                    if i+1<len(pr):
                        n=pr[i+1]
                        Fo[A]|=(first(n)-{'e'}) if n.isupper() else {n}
                        if n.isupper() and 'e' in first(n): Fo[A]|=follow(X)
                    elif X!=A: Fo[A]|=follow(X)
    return Fo[A]

for A in p: print("first("+A+") =",sorted(first(A)))
for A in p: print("follow("+A+") =",sorted(follow(A)))
```
