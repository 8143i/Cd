```python
import ply.lex as lex
import sys

# Token list
tokens = (
    'IMPORTS', 'STRING', 'KEYWORD', 'FUNCTION',
    'FLOAT', 'INT', 'OPERATOR', 'ID',
    'LPARAN', 'RPARAN', 'SEPARATOR',
    'LBRACE', 'RBRACE'
)

# Token regex rules
t_IMPORTS   = r'<stdio.h>|<conio.h>|<stdlib.h>'
t_STRING    = r'\".*\"'
t_KEYWORD   = r'\#include|if|else|for|break|int|float|void|String|char|double|while|do'
t_FUNCTION  = r'printf|scanf|clrscr|getch'
t_FLOAT     = r'\d+\.\d+'
t_INT       = r'\d+'
t_OPERATOR  = r'\+|-|\*|/|=|==|<|>|>=|<='
t_ID        = r'[a-zA-Z_][a-zA-Z0-9_]*'
t_LPARAN    = r'\('
t_RPARAN    = r'\)'
t_SEPARATOR = r'[;:,]'
t_LBRACE    = r'\{'
t_RBRACE    = r'\}'

# Ignore spaces and tabs
t_ignore = ' \t'

# Handle newlines
def t_NEWLINE(t):
    r'\n+'
    t.lexer.lineno += len(t.value)

# Handle errors
def t_error(t):
    print("Illegal character:", t.value[0])
    t.lexer.skip(1)

# ✅ Build lexer safely for Jupyter
lexer = lex.lex(module=sys.modules[__name__], optimize=1, nowarn=True)

# Read and tokenize input file
with open("text.cpp") as f:
    lexer.input(f.read())

# Print tokens
for tok in lexer:
    print(tok)
```
