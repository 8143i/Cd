```python
import re

p = r'''
(?P<IMPORTS>   <stdio.h> | <conio.h> | <stdlib.h> ) |
(?P<STRING>    ".*?" ) |
(?P<KEYWORD>   #include | if | else | for | break | int | float | void | String | char | double ) |
(?P<FUNCTION>  printf | scanf | clrscr | getch ) |
(?P<FLOAT>     \d+\.\d+ ) |
(?P<INT>       \d+ ) |
(?P<OPERATOR>  \+\+ | -- | == | = | \+ | - | \* | / | < | > ) |
(?P<LPARAN>    \( ) |
(?P<RPARAN>    \) ) |
(?P<SEPARATOR> [;:,] ) |
(?P<LBRACE>    \{ ) |
(?P<RBRACE>    \} ) |
(?P<ID>        [a-zA-Z_][a-zA-Z0-9_]* )
'''
code = open('text.cpp').read()

for m in re.finditer(p, code):
    print(f'{m.group()} -> {m.lastgroup}')
```
