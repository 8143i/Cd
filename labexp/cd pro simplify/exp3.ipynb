```python
e = 'ε'

def remove(p):
    res = []
    for A, prods in p:
        alpha = [r[1:] for r in prods if r.startswith(A)]
        beta  = [r for r in prods if not r.startswith(A)]

        if alpha:
            res.append((A, [b + A + "'" for b in beta]))
            res.append((A+"'", [a + A + "'" for a in alpha] + [e]))
        else:
            res.append((A, prods))
    return res


# Input
p = []
for _ in range(int(input("No of productions: "))):
    A, rhs = input().split("->")
    p.append((A.strip(), [x.strip() for x in rhs.split("|")]))

print("\nOriginal:")
for A, r in p:
    print(A, "->", " | ".join(r))

p = remove(p)

print("\nAfter Removing Left Recursion:")

# First print original non-terminals
for A, r in p:
    if "'" not in A:
        print(A, "->", " | ".join(r))

# Then print new non-terminals (with ')
for A, r in p:
    if "'" in A:
        print(A, "->", " | ".join(r))
```
